# ISIC 2017 - Skin Lesion Segmentation 

```python
# pip install image-dataset-viz
from image_dataset_viz import render_datapoint

import PIL
from PIL import Image
```

```python
image = Image.open('data/skin/images/ISIC_0000018.jpg')
image = image.convert('RGB')
```

```python
mask = Image.open('data/skin/masks/ISIC_0000018_segmentation.png')
background = Image.new('L', size=image.size)
g = Image.merge('RGB', (background, mask.convert('L'), background))
```

```python
blended = render_datapoint(image, g, blend_alpha=0.2)
```

```python
print(blended.size)
```
